# To create an undirected graph and add nodes and edges to a graph

## Import networkx and matplotlib.pyplot in the project file.

```python
import networkx as nx
import matplotlib.pyplot as plt
```

## To create an empty undirected graph

```python
G = networkx.Graph()
```

## To add a node

```python
G.add_node(1)
G.add_node(2)
G.add_node(3)
G.add_node(4)
G.add_node(7)
G.add_node(9)
```

## To add an edge

```python
G.add_edge(1,2)
G.add_edge(3,1)
G.add_edge(2,4)
G.add_edge(4,1)
G.add_edge(9,1)
G.add_edge(1,7)
G.add_edge(2,9)
```

```python
nx.draw(G)
```

## To add numbering in the node add one argument with_labels=True in draw() function.  

```python
nx.draw(G, with_labels = True)
```

## To get all the nodes, edges of a graph

```python
# To get all the nodes of a graph
node_list = G.nodes()

print(node_list)
  
# To get all the edges of a graph
edge_list = G.edges()

print(edge_list)
```

## To remove a node, edge of a graph

```python
# To remove a node of a graph
G.remove_node(4)
node_list = G.nodes()

print(node_list)
  
# To remove an edge of a graph
G.remove_edge(1,2)
edge_list = G.edges()

print(edge_list)

nx.draw(G, with_labels = True)
```

## To find number of nodes, edges

```python
# To find number of nodes
n = G.number_of_nodes()
print(n)
  
# To find number of edges
m = G.number_of_edges()
print(m)
```

# Creating Weighted undirected Graph

```python
import networkx as nx
G = nx.Graph()
  
edges = [(1, 2, 19), (1, 6, 15), (2, 3, 6), (2, 4, 10), 
         (2, 6, 22), (3, 4, 51), (3, 5, 14), (4, 8, 20),
         (4, 9, 42), (6, 7, 30)]
G.add_weighted_edges_from(edges)
#nx.draw_networkx(G)
nx.draw(G, with_labels = True)
```

## We can add the edges via an Edge List, which needs to be saved in a .txt format (eg. edge_list.txt)

```python
G = nx.read_edgelist('edge_list.txt', data =[('Weight', int)])
#nx.draw(G, with_labels = True)
#nx.draw(G, pos=nx.spiral_layout(G))
nx.draw(G, pos=nx.shell_layout(G),with_labels = True)
#nx.draw(G, pos=nx.spiral_layout(G),with_labels = True)
plt.show()
```

## Edge list can also be read via a Pandas Dataframe

```python
import pandas as pd
  
df = pd.read_csv('edge_list.txt', delim_whitespace = True, 
                   header = None, names =['n1', 'n2', 'weight'])
  
G = nx.convert_matrix.from_pandas_edgelist(df, 'n1', 'n2', edge_attr ='weight')
  

print(list(G.edges(data = True)))
```

# Creating Directed Graph

```python
import networkx as nx
G = nx.DiGraph()
G.add_edges_from([(1, 2), (1, 4), 
                  (2, 3), (2, 4), 
                  (3, 5), 
                  (4, 3), 
                  (5, 4), (5, 1)])
  
plt.figure(figsize =(9, 9))
nx.draw_networkx(G, with_labels = True, node_color ='red')
  
# getting different graph attributes
print("Total number of nodes: ", int(G.number_of_nodes()))
print("Total number of edges: ", int(G.number_of_edges()))
print("List of all nodes: ", list(G.nodes()))
print("List of all edges: ", list(G.edges()))
print("In-degree for all nodes: ", dict(G.in_degree()))
print("Out degree for all nodes: ", dict(G.out_degree))

  
```
